- In machine learing we have one important property is **Independent of variable

- Machine learning models expects variables  are independent each other

- One hot coder creating new columns,the new columns are independents each other

- indepents variable means both variables have 90 degrees phase shift

- 90 degree phase shift also called as perpendicular each other

- perpendicular means orthogonal each other

- if number of unique lables increase,new columns also increase

-  the more columns leads data is more

- complexity increase  more computation power required
